```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = r'C:\Users\Admin\Desktop\Chinmayee Verma\Breast Cancer Wisconsin EDA\data.csv'
data = pd.read_csv(data_path)

print(data.head())
```

```python
data_cleaned = data.drop(columns=['Unnamed: 32'])
summary_stats = data_cleaned.describe()
print(data_cleaned.head())
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(20, 15))
plt.show()
```

```python
plt.subplot(1, 1, 1)
size_features = ['radius_mean', 'perimeter_mean', 'area_mean']
sns.scatterplot(data=data, x='radius_mean', y='area_mean')
plt.title('Radius vs Area (Size Relationship)')
plt.xlabel('Radius Mean')
plt.ylabel('Area Mean')
plt.show()
```

```python
plt.subplot(1, 1, 1)
sns.scatterplot(data=data, x='texture_mean', y='smoothness_mean')
plt.title('Texture vs Smoothness')
plt.xlabel('Texture Mean')
plt.ylabel('Smoothness Mean')
```

```python
data_cleaned.hist(bins=20, figsize=(20,16))
plt.show()
```

```python

numeric_data = data_cleaned.select_dtypes(include=[float, int])

plt.figure(figsize=(15, 10))
corr_matrix = numeric_data.corr()

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
correlation_matrix = data.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

```python
import sklearn
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_cleaned.drop(columns=['id', 'diagnosis']))
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=pca_result[:,0], y=pca_result[:,1], hue=data_cleaned['diagnosis'], palette='Set1')
plt.title("PCA of Breast Cancer Data")
plt.show()
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=pca_result[:,0], y=pca_result[:,1], hue=data_cleaned['diagnosis'], palette='Set2')
plt.title("PCA of Breast Cancer Data")
plt.show()
explained_variance = pca.explained_variance_ratio_
explained_variance
```
